```python
import psycopg2

def connect_to_postgresql():
    try:
        # Connect to the PostgreSQL database (replace the placeholders with your own credentials)
        conn = psycopg2.connect(
            host="127.0.0.1",
            database="testdb",
            user="postgres",
            password="root",
            port=5432
        )

        if conn:
            print("Connected to PostgreSQL database.")

            # Create a cursor object to interact with the database
            cursor = conn.cursor()

            # Execute a simple query
            query = "SELECT * FROM your_table"
            cursor.execute(query)

            # Fetch and print the query results
            rows = cursor.fetchall()
            for row in rows:
                print(row)

            # Close the cursor and connection
            cursor.close()
            conn.close()
            print("Connection closed.")
        else:
            print("Failed to connect to PostgreSQL database.")

    except psycopg2.Error as e:
        print(f"Error occurred: {e}")

# Example usage
connect_to_postgresql()
```

```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine

# Step 1: Read Excel file into a pandas DataFrame
excel_file = 'ESD.xlsx'  # Replace 'your_excel_file.xlsx' with the path to your Excel file
df = pd.read_excel(excel_file)

# Step 2: Establish a connection to your PostgreSQL database
DB_HOST = 'localhost'  # Replace 'your_host' with your PostgreSQL host
DB_NAME = 'testdb'  # Replace 'your_database' with your PostgreSQL database name
DB_USER = 'postgres'  # Replace 'your_username' with your PostgreSQL username
DB_PASSWORD = 'root'  # Replace 'your_password' with your PostgreSQL password

# Construct the database URL
DB_URL = f'postgresql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}/{DB_NAME}'

# Create an engine to connect to the database
engine = create_engine(DB_URL)

# Step 3: Insert data from DataFrame into PostgreSQL table
table_name = 'ESD'  # Replace 'your_table_name' with the name of your PostgreSQL table

try:
    df.to_sql(table_name, engine, index=False, if_exists='append')
    print("Data successfully pushed to PostgreSQL!")
except Exception as e:
    print("Error:", e)
```
