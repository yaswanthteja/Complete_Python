```python
class Solution:
    #@ param A: list of integers
    #@ return an integer
    def solve(self,A):
        min_element=min(A)
        max_element=max(A)
        count=0
        for i in range(len(A)):
            if A[i]==min_element or A[i]==max_element:
                count+=1
        return (len(A)-count)
obj=Solution()
A=[2,3,4,5,6,7,8]
obj.solve(A)
```
