### What is matplotlib?
A matplotlib is a python library usedd for creating static, animated, and interactive plots

 Data visualization : It's the graphical representation of data to identify patterns , trends and insights.

```python
# import library
import matplotlib.pyplot as plt

# pyplot  is a collection of functions that make matplotlib work as MATLAB
```

```python
# sample data
x=[1,2,3,4,5]
y=[10,20,25,35,40]

plt.plot(x,y) # line chart
plt.show()  # display chart in the jupyter notebook
```

```python

# 2. customize charts


plt.figure(figsize=(8,4)) # figuresize in inches
plt.plot(x,y, color='green',marker='o',linestyle='dashed',linewidth=2, markersize=12)

plt.title('Data ploted in line chart') # title for chart
plt.xlabel("X-axis") # title for x axis
plt.ylabel("Y-axis") # title for y axis
plt.show()
```

```python
# 3. adavanced line chart - Multiple lines & legends
# data
x=[1,2,3,4,5]
y1=[10,20,25,35,45]
y2=[20,30,35,45,55]


plt.plot(x, y1, label='sales of 2024')
plt.plot(x, y2, label='sales of 2025')

plt.title("YOY sales")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.legend() # show legends
plt.show()

```

```python
# Bar charts
x=['A','B','C','D','E']
y=[10,20,55,35,45]

plt.bar(x,y)

plt.xlabel("categories on x axis")
plt.ylabel("category values on y axis")
plt.title("Bar chart")

plt.show()

```

```python
# Histogram
# used for distribution analysis

# data
import random
data=[random.randint(1,20) for _ in range(500)]

# histogram
plt.hist(data,bins=20,histtype='bar') #bins no of hist  histtype 'bar', 'barstacked', 'step', 'stepfilled'

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Histogram")


plt.show()
```

```python
# pie chart
# used to show part to whole relationships/data

# data

categories=['A','B','C','D','E']
sales=[10,20,55,35,45]

# pie chart
plt.pie(sales, labels=categories, autopct='%1.1f') # autopct -> display percentages and 1.1f display percentage

plt.title("Sales by categories using pie chart ")

plt.show()
```

```python
# 7. scatterplot
# used to find relationships between variables

# data
y1=[10,20,25,35,45]
y2=[20,30,35,45,55]

plt.scatter(y1,y2)

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter plot")

plt.show()
```

```python
# 8. subplots
# used to show  multiple charts in one frame

categories=['Mon','Tue','Wed','Thu','Fri']
sales=[10,20,55,35,45]

# data 2
y1=[10,20,25,35,45]
y2=[20,30,35,45,55]


plt.figure(figsize=(10,4)) # figure size in inches

# first plot bar chart
plt.subplot(1,2,1) # 1: row ,2: column,1: postion

plt.bar(categories,sales)
plt.xlabel("week Days")
plt.ylabel(" sales")
plt.title("Daily sales")


# second plot -scatter chart

plt.subplot(1,2,2) # 1: row ,2: column,2: postion

plt.scatter(y1,y2)
plt.xlabel("user 1")
plt.ylabel(" user 2")
plt.title("user example")
plt.show()
```

```python
#9. Matplotlib with pandas -real data

import pandas as pd

# create data frame

data={
    'month':['Jan','Feb','Mar','Apr','May','Jun'],
    'sales':[12000,11000,10000,9000,8000,17000]
}

df=pd.DataFrame(data)
df
```

```python
# plot

plt.bar(df['month'],df['sales'])
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Matplotlib with pandas Monthly sales ")
plt.show()
```

```python
# 10 save charts

plt.bar(df['month'],df['sales'])
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Matplotlib with pandas Monthly sales ")

plt.savefig("Monthly_sales_from_df.png") # used to save images

plt.show()
```
