```python
L=[10,20,30,40]
iter_obj=iter(L)  # iterator
# access the elements 

print(next(iter_obj)) #10
print(next(iter_obj)) #20 
print(next(iter_obj)) #30
print(next(iter_obj)) #40
```

- Here next() function is used to get the next item from the iterator. 
- The first call to next(it) retrieves the first item (1) from the iterator.
- The second call to next(it) retrieves the second item (2) from the iterator.
- The third call to next(it) retrieves the third item (3) from the iterator.
- The fourth call to next(it) retrieves the fourth item (4) from the iterator.
- every time next() is called, it returns the next item in the sequence until there are no more items left.
- When there are no more items to return, a StopIteration exception is raised.



- next() remembers the state of the iterator, so each call to next() continues from where the last call left off.
- This is why each call to next() returns the subsequent item in the sequence.
- If you call next() again after all items have been retrieved, it will raise a StopIteration exception, indicating that there are no more items to iterate over.




```python
## next() knows the current iteration

L=[10,20,30,40]

iter_obj=iter(L) # iterator

print(next(iter_obj)) #10
print(next(iter_obj)) # 20

for ele in iter_obj:
    print(ele)       #30
                     # 40
```

```python
# Making our own for loop

L=[10,20,30,40]

def for_loop(iterable):
    iterator=iter(iterable)
    try:
        while True:
            print(next(iterator))
    except StopIteration:
        pass
for_loop(L)


```

```python

# using recurssion
def for_loop(iterable):
    iterator = iter(iterable)
    
    def loop():
        try:
            print(next(iterator))
            loop()  # recursive call
        except StopIteration:
            pass

    loop()

# Example
L = [10, 20, 30, 40]
for_loop(L)
```

```python
## using Custom iterator creating our own range() 

class Myrange:
    def __init__(self, start, end):
        self.current = start
        self.end = end

    def __iter__(self):
        return self

    def __next__(self):
        if self.current >= self.end:
            raise StopIteration
        current = self.current
        self.current += 1
        return current


my_iter = Myrange(1, 4)

for value in my_iter:
    print(value)
```

```python
# custom range() with step
class MyRange:
    def __init__(self,start,end,step=1):
        self.current=start
        self.end=end
        self.step=step

    def __iter__(self):
        return self
    
    def __next__(self):
        if (self.step >0 and self.current>=self.end) or (self.step<0 and self.current<=self.end):
            raise StopIteration
        value=self.current
        self.current+=self.step
        return value
    
for num in MyRange(1,10,2):
    print(num)
```

## Generators

A generator is a function that returns an iterator object. Instead of computing and storing all values at once (like a list), it yields values one by one â€” pausing after each and resuming when needed.

```python

# Using Iterator
class Counter:
    def __init__(self, max):
        self.max = max
        self.current = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.max:
            raise StopIteration
        else:
            self.current += 1
            return self.current - 1

for num in Counter(5):
    print(num)
```

```python
# same using generator

# Simplified version using yield
def counter(max):
    current = 1
    while current <= max:
        yield current
        current += 1

# Using the generator
for num in counter(5): # here we can give any large number then also it gives us the output cause yield doesn't store in the memory it give on the go
    print(num)
```

```python
# instead of yield 

def counter(n):
    numbers=[]
    count=1
    while count<=n:
        numbers.append(count)
        count+=1
    return numbers   # cause return try to give result at once 
n=int(input("Enter a number: "))    ## memory error  if we try to give large number
for num in counter(n):
    print(num)
```
