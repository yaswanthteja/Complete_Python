If a function only prints and has no `return`, calling it directly will  **only show the printed output** .
If you store its result and print that, you'll see the printed output **plus **`None`.

```python
def greet(wish):
    print(wish)

greet("hello")

```

If you store its result and print that, you'll see the printed output **plus **`None`.

```python
def greet(wish):
    print(wish)


print(greet("hello"))
```

```python
def greet(wish):
    print(wish)

result=greet("hello")

print(result)
```
