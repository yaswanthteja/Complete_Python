```python
# set={} unordered and immutable but can add or remove elements but no duplicates are allowed
```

```python
fruits={"Apple","orange","banana","coconut"}
```

```python
type(fruits)
```

```python
# To display all the attributes and Methods of a set we can use the dir()

print(dir(fruits))
```

```python
# To find out the indepth descrption of these we can use help()
print(help(fruits))
```

```python
# To find out the length of the set we use len

print(len(fruits))
```

```python
# to find out that a value is present in the set 

print('pineapple' in fruits)
```

```python
# we can't use indexing cause they are un ordered
print(fruits[0])
```

```python
# we can't change the elements of set but we can add or remove elements of the set

fruits.add("pineapple")

print(fruits)
```

```python
# we can remove an element

fruits.remove("Apple")
print(fruits)
```

```python
# We can use pop method to remove the first element

fruits.pop()
print(fruits)
```

```python
# To clear the elements we clear method

fruits.clear()
print(fruits)
```
