## Loop in list

```python
fruits = ['orange', 'apple', 'pear']

for i in fruits:
    print(i)
```

```python
fruits = ['orange', 'apple', 'pear']

for i in range(len(fruits)):
    print(i,fruits[i])
```

### enumarate() to generate both index and value during iteration

```python
fruits = ['orange', 'apple', 'pear']

for i,value in enumerate(fruits):
    print(i,value)
```

```python
# starting index from 1
students=['Appu','fil','zill']

for i in range(len(students)):
    print(i+1,students[i])
```

```python
# using while loop
ruits = ['orange', 'apple', 'pear']

i = 0
while i < len(fruits):
    print(fruits[i])
    i+=1
```

```python
fruits=['apple', 'orange','mangaos','malberry']

i=0
while i<len(fruits):
    print(i+1,fruits[i])
    i+=1
```

### we use zip() to iterate through two lists

```python
# we use zip() to iterate through two lists
fruits=['apple', 'orange','mangaos','malberry']
prices=[10,5,8,4]

for f,p in zip(fruits,prices):
    print(f,p)
```

### methods of list

```python
# count
cars = ['Toyota','Chevrolet','Renault','Toyota']

print(cars.count('Toyota'))
```

```python
# sort
numbers = [5,2,7,1,3]
numbers.sort()
print(numbers)
```

```python
alphabates = ['A','F','B','K','D']
alphabates.sort()
print(alphabates)
```

```python
# descending order

num2 = [1,5,2,6,8]
num2.sort(reverse=True)
print(num2)  # output: [8, 6, 5, 2, 1]

alphabates2 = ['A','F','B','K','D']
alphabates2.sort(reverse=True)
print(alphabates2) # output: ['K', 'F', 'D', 'B', 'A']
```

```python
# sorted()
# ascending order
numbers = [5,2,7,1,3]
rev_num = sorted(numbers)
print(rev_num)

# descending order
num2 = [1,8,5,9,3]
rev_num2 = sorted(num2,reverse=True)
print(rev_num2)
```

```python
# clear - clear entire list
num = [1,2,3,4,5]
num.clear()
print(num)
```

```python
# copy

l1 = [2,3,5,7,11,13]
l2 = l1.copy()

print(l2)
```

```python
# we can use * to combine lists

a=[1,2]
b=[3,4]

ls=[*a,*b]
print(ls)
```

```python
# we can use * to unpack lists

dog=['rocky',2,'chicken','mutton','prawns']

name,age, *fav_food=dog

print(name)
print(age)
print(fav_food)
```

## list comprehension

```python

squares = [x ** 2 for x in range(1, 6)]  # Create a list of squares
squares
```

```python
my_list = [45,12,67,8,2,6,90,13,15]

my_list2 = [i for i in my_list if i%2 == 0]

print(my_list2)
```

```python
mixed_list = ['Apple','Banana',12,15,7,2,3,'Orange','Mango']

# if type is equal to int then append square of the number
# otherwise append first character of string

mixed_list2 = [i**2 if type(i) == int else i[0] for i in mixed_list]
print(mixed_list2)
```
