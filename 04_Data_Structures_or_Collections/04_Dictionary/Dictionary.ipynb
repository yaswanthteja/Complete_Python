dictionary = a collection of {key:value} pairs ordered and Changeable . no duplicates (keys)


```python
capitals={
    "USA":"washington D.C",
    "India":"New Delhi",
    "China": "Beijing",
    "Russia" :"Moscow"
}
```

```python
# if we want to see different attributes and methods of dictionary dir()

print(dir(capitals))
```

```python
# if we want indepth explanation of this attributes and methods we can use help
print(help(capitals))
```

```python
# Methods
```

```python
capitals['USA']
```

```python
# or We can also use get method using key to get value of that particular key

print(capitals.get('USA'))
```

```python
# if the key is not present in dictionary it will return None
print(capitals.get('Japan'))
```

```python
# we can also use if else to see if the key present in dicitionary
if capitals.get("Japan"):
    print("that capital exists")
else:
    print("that capital doesn't exist")
```

```python
# updating the dictionary using update method 
# by using the update method we can add a new value pair to dictionary or update existing value pair

capitals.update({'Germany':'Berlin'})
print(capitals)
```

```python
# updating existing key value

capitals.update({'USA':"W.DC"})
print(capitals)
```

```python
# removing a key value pair using pop method

capitals.pop('China')
capitals
```

```python
# we remove recent key value pair by using popitem method

capitals.popitem()
```

```python
# to get all keys froms dictionary
capitals.keys()
```

```python
# so it is in object form ,by using for loop we can get only keys

for key in capitals.keys():
    print(key)
```

```python
# like wise we also  have values methods

capitals.values()
```

```python
# so it is also in object from we can use for loop

for value in capitals.values():
    print(value)
```

```python
# By using items method returns  a dictionary object which resembled a 2d list of tuples 
capitals.items()
```

```python
# we can iterate this using a for loop
for key,value in capitals.items():
    print(key,value)
```

```python
# to clear the dictionary we use clear method 
capitals.clear()
print(capitals)
```
